### Tensors
- Generally speaking, Tensors are n-dimensional arrays with a specified data type. 
    - That is, each component of the Tensors has the same data type (e.g., int32 or float32) 
    - Various methods can create tensors, two of which–that is–**constants and variables** are the most common ones.

__Tensor의 속성__
- 타입(Type): `string`, `float32`, `float16`, `int32`, `int8` 등
- 형상(Shape): 0, 1, 2차원 등의 데이터 차원 
- 축(Rank): 차원의 개수

### 변수(Variables)

Variables hold a persistant shared state across your computation. The most common use case of Variables is the model's trainable parameters.

- The only way to create variables is to use `tf.Variable(<required-initial-value>, name=<optional-name>)` class.
    - Tensorflow uses the `initial-value` to infer the shape and the type of the variable. 
    - The shape and the type of variable, once specified, cannot be changed during the computation.  
    - Tensorflow cleans up variables when the runtime changes its scope and the variable is not referenced anymore. 
    - Therefore, it is your responsibility to keep track of variables in your Tensorflow program (Good news: Tensorflow's high-level APIs handles that automatically)

```python
w = tf.Variable(20., name="my_var01")
print('w =', w)

initializer = tf.initializers.GlorotUniform()
x = tf.Variable(initializer(shape=(2, 5)), name="my_var02")
print('\nx =', x)

y = tf.Variable(tf.zeros([5]), name='my_var03')
print('\ny =', y)

z= tf.Variable([[1.,2,3],
                [1,1,1]])
print('\nz =', z)
```

__Variables' APIs are mostly similar to Tensors. Hence, we can treat them like a standard Tensor.__

```python
v = w + 1.  # v is a tf.Tensor and is calculated as the result of
            # a mathematical expression that is based on a variable(w).
            # tf.Variable gets automatically converted to a tf.Tensor 
            # representing its value when it is envolved in a expression.
print("v =", v)
print(f"v's type = {type(v)}")
print(f"w's type = {type(w)}")
```

To change the variable's current value, you can use methods such as `assign` and `assign_add`. ([More info](https://www.tensorflow.org/api_docs/python/tf/Variable))

```python
w.assign(v)
print('w =', w)
w.assign_add(v)
print('w =', w)
```

A convenient way to think about and visualize computations in TensorFlow is in terms of graphs. Let's define a simple function in TensorFlow to construct this computation function:

```python
a = tf.constant(15)
b = tf.constant(61)

def func(a,b):
    c = a + b
    d = b - 1
    e = c*d
    return e

print(func(a,b))
```

```python
# tf.transpose(t, perm) permutes the dimensions according to the `perm` parameter.
t1 = tf.constant([[1, 2, 3], [4, 5, 6]]) # (2,3) -> (3, 2)
print("tf.transpose(t1, [1, 0]) =", tf.transpose(t1, perm=[1, 0])) 
```

```python
# It also works in higher dimensions
t1 = tf.constant(np.arange(1, 41, dtype=np.int32), shape=[2, 5, 4])
print(f"\nt1 =\n{t1}\n")

t1_t = tf.transpose(t1, perm=[0, 2, 1])
print(f"\nt1_t.shape = {t1_t.shape}\n")
print(f"\nt1_t =\n {t1_t}\n")

# You can permute the order of more than two dimensions at the same time.
t1_t = tf.transpose(t1, perm=[2, 0, 1])
print(f"\nt1_t.shape = {t1_t.shape}\n")
print(f"\nt1_t = \n{t1_t}\n")
```

**Reshaping**
You can create a new tensor from an existing tensor with different shape but same values. The only rule is that __the new tensor's size should be equal to that of the previous one__.

```python
# Examples from https://www.tensorflow.org/api_docs/python/tf/reshape

t = tf.constant([1, 2, 3, 4, 5, 6, 7, 8, 9]) # [9]
print(f"t_new = tf.reshape(t, [3, 3]); t_new => \n {tf.reshape(t, [3, 3])}")
```

```python
t = tf.constant([[[1, 1], [2, 2]],
                [[3, 3], [4, 4]]]) # [2, 2, 2]
print(f"\nt_new = tf.reshape(t, [2, 4]); t_new => \n {tf.reshape(t, [2, 4])}")
```

__-1 can also be used to automatically calculate the shape.__

```python
t = tf.constant([[[1, 1, 1],
                 [2, 2, 2]],
                [[3, 3, 3],
                 [4, 4, 4]],
                [[5, 5, 5],
                 [6, 6, 6]]]) # [3, 2, 3]

# -1 => 18
print(f"\nt_new = tf.reshape(t, [-1]); t_new => \n {tf.reshape(t, [-1])}")
# -1 => 9
print(f"\nt_new = tf.reshape(t, [2, -1]); t_new => \n {tf.reshape(t, [2, -1])}")
# -1 => 2
print(f"\nt_new = tf.reshape(t, [-1, 9]); t_new => \n {tf.reshape(t, [-1, 9])}")
# -1 => 3
print(f"\nt_new = tf.reshape(t, [2, -1, 3]); t_new => \n {tf.reshape(t, [2, -1, 3])}")
```

```python
# Convert to a scalar using shape `[]`
t = tf.constant([5])
t_scalar = tf.reshape(t, [])
print('t',t,'\n')
print('t',t_scalar,'\n')
```

**Advanced Reshaping**

```python
# tf.tile(t, multiples) creates a new tensor by replicating `t` `multiples` times.
t = tf.constant([1, 2, 3, 4]) # [4]
print(f"tf.tile(t, [2]) = \n{tf.tile(t, [2])}") # [8]

t = tf.constant([[1, 2, 3, 4]]) # [1, 4]
print(f"\ntf.tile(t, [3, 1]) = \n{tf.tile(t, [3, 1])}") # [3, 4]
```

**Combining Tensors**

```python
t1 = tf.constant([[1, 1, 1], [1, 1, 1]]) # [2, 3]
t2 = tf.constant([[2, 2, 2], [2, 2, 2]]) # [2, 3]
t3 = tf.constant([[3, 3, 3], [3, 3, 3]]) # [2, 3]

print(f"tf.concat([t1, t2, t3], axis=0) = \n{tf.concat([t1, t2, t3], axis=0)}") # [6, 3]
print(f"\ntf.concat([t1, t2, t3], axis=1) = \n{tf.concat([t1, t2, t3], axis=1)}") # [2, 9]
```

```python
t1 = tf.constant([1, 1, 1, 1])
t2 = tf.constant([2, 2, 2, 2])
t3 = tf.constant([3, 3, 3, 3])

# Vertical Stack
t_stack_1 = tf.stack([t1, t2, t3], axis=0)
print(f"tf.stack([t1, t2, t3], axis=0) = \n{t_stack_1}")
print('Shape:', t_stack_1.shape.as_list())

# Horizontal Stack
t_stack_2 = tf.stack([t1, t2, t3], axis=1)
print(f"tf.stack([t1, t2, t3], axis=0) = \n{t_stack_2}")
print('Shape:', t_stack_2.shape.as_list())
```

**Reducing**

In python, the reduce(fun,seq) function is used to apply a particular function passed in its argument to all of the list elements mentioned in the sequence passed along. This function is defined in “functools” module.

```python
# python code to demonstrate working of reduce()

# importing functools for reduce()
import functools

# initializing list
list = [ 1 , 3, 5, 6, 2, ]

# using reduce to compute sum of list
print ("The sum of the list elements is : ",end="")
print (functools.reduce(lambda a,b : a+b,list))

# using reduce to compute maximum element from list
print ("The maximum element of the list is : ",end="")
print (functools.reduce(lambda a,b : a if a > b else b,list))
```
